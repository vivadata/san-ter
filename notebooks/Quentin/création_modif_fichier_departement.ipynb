### JOINTURE PATHOLOGIE + JSON DEPARTEMENT

```python
import geopandas as gpd
import pandas as pd

# Charger le GeoJSON des départements (100m)
gdf_departements = gpd.read_file("departements-100m.geojson")

# Charger le CSV pathologie par département
df_pathologie = pd.read_csv("pathologie par departement.csv")

# Exemple : df_pathologie contient les colonnes : 'departement', 'Famille Pathologie', 'valeur' ou 'indice'
print(df_pathologie.head())
```

```python
gdf_departements['code_departement'] = gdf_departements['code'].astype(str)
df_pathologie['ZONE_0'] = df_pathologie['ZONE_0'].astype(str)
```

### JOINTURE

```python
gdf_pathologie = gdf_departements.merge(
    df_pathologie,
    left_on='code_departement',
    right_on='ZONE_0',
    how='left'
)

# Sélectionner uniquement les colonnes utiles
gdf_pathologie_reduit = gdf_pathologie[['code_departement', 'nom', 'geometry', 'FAMILLE_PATHOLOGIE', 'ind_freq_sum']]

# Vérification
print(gdf_pathologie_reduit.head())
```

```python
gdf_pathologie_reduit.to_file("FV_pathologie_departements.gpkg", driver="GPKG")

gdf_pathologie_reduit['geometry'] = gdf_pathologie_reduit['geometry'].apply(lambda x: x.wkt)

gdf_pathologie_reduit.to_csv("FV_pathologie_departements.csv", index=False)
```
